```python
# Daily Balance OG Approach using event log

sql ="""
-- Create or Replace Table: daily_token_balances

--CREATE OR REPLACE TABLE your_schema.daily_token_balances AS
--Universe of all transfers for arb
WITH erc20_transfers AS (
    SELECT
        el.block_number,
        el.block_timestamp,
        el.contract_address AS token_address,
        el.topic_1 AS from_address,
        el.topic_2 AS to_address,
        CAST(el.data AS NUMBER) AS amount_raw
    FROM arbitrum.core.fact_event_logs el
    JOIN arbitrum.core.dim_contracts tc 
        ON el.contract_address = tc.address
    WHERE el.topic_0 = '0xddf252ad1be2c89b69c2b068fc378daa952ba7f163c4a11628f55a4df523b3ef'--Only Transfer transactions
        AND el.contract_address = lower('0x912CE59144191C1204E64559FE8253a0e49E6548')
),

-- Split transactions for from and to address, making from address amount values negative, to represent value leaving wallet
all_transfers AS (
    -- From address therefore negative amount_raw
    SELECT 
        DATE_TRUNC('day', block_timestamp) AS date,
        token_address,
        from_address AS wallet_address,
        -amount_raw AS delta_raw
    FROM erc20_transfers
    UNION ALL
    -- To address therefore positive amount_raw
    SELECT 
        DATE_TRUNC('day', block_timestamp) AS date,
        token_address,
        to_address AS wallet_address,
        amount_raw AS delta_raw
    FROM erc20_transfers
),

--Calculate daily balance amount
daily_deltas AS (
    SELECT
        wallet_address,
        token_address,
        date,
        SUM(delta_raw) AS daily_delta_raw
    FROM all_transfers
    GROUP BY wallet_address, token_address, date
),

-- Rolling balance over time
cumulative_balances AS (
    SELECT
        wallet_address,
        token_address,
        date,
        SUM(daily_delta_raw) OVER (
            PARTITION BY wallet_address, token_address
            ORDER BY date
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        ) AS balance_raw
    FROM daily_deltas
)

SELECT
    cb.wallet_address,
    cb.token_address,
    tc.symbol AS token_symbol,
    tc.decimals,
    cb.date,
    cb.balance_raw,
    cb.balance_raw / POWER(10, tc.decimals) AS balance_adjusted
FROM cumulative_balances cb
JOIN arbitrum.core.dim_contracts tc 
    ON cb.token_address = tc.address
WHERE cb.balance_raw != 0; -- Filter out 0 balances
"""
token_balances = flipside.query(sql)

#Format and create dataframe from flipside returned query
columns = token_balances.columns
rows = token_balances.rows
token_balance_logs_df = pd.DataFrame(rows, columns=columns)

# see df head
display(token_balance_logs_df.head())
```

```python
# Daily Balance Using ez _token_transfers

sql ="""
-- Create or Replace Table: daily_token_balances

--CREATE OR REPLACE TABLE your_schema.daily_token_balances AS
--Universe of all transfers for arb
WITH erc20_transfers AS (
    SELECT
        tt.block_number,
        tt.block_timestamp,
        tt.contract_address AS token_address,
        tt.from_address,
        tt.to_address,
        tt.amount AS amount_raw
    FROM arbitrum.core.ez_token_transfers tt
    JOIN arbitrum.core.dim_contracts tc 
        ON tt.contract_address = tc.address
    WHERE tt.contract_address = lower('0x912CE59144191C1204E64559FE8253a0e49E6548')
        and tt.amount > 0
),

-- Split transactions for from and to address, making from address amount values negative, to represent value leaving wallet
all_transfers AS (
    -- From address therefore negative amount_raw
    SELECT 
        DATE_TRUNC('day', block_timestamp) AS date,
        token_address,
        from_address AS wallet_address,
        -amount_raw AS delta_raw
    FROM erc20_transfers
    UNION ALL
    -- To address therefore positive amount_raw
    SELECT 
        DATE_TRUNC('day', block_timestamp) AS date,
        token_address,
        to_address AS wallet_address,
        amount_raw AS delta_raw
    FROM erc20_transfers
),

--Calculate daily balance amount
daily_deltas AS (
    SELECT
        wallet_address,
        token_address,
        date,
        SUM(delta_raw) AS daily_delta_raw
    FROM all_transfers
    GROUP BY wallet_address, token_address, date
),

-- Rolling balance over time
cumulative_balances AS (
    SELECT
        wallet_address,
        token_address,
        date,
        SUM(daily_delta_raw) OVER (
            PARTITION BY wallet_address, token_address
            ORDER BY date
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        ) AS balance_raw
    FROM daily_deltas
)

SELECT
    cb.wallet_address,
    cb.token_address,
    tc.symbol AS token_symbol,
    tc.decimals,
    cb.date,
    cb.balance_raw,
    cb.balance_raw / POWER(10, tc.decimals) AS balance_adjusted
FROM cumulative_balances cb
JOIN arbitrum.core.dim_contracts tc 
    ON cb.token_address = tc.address
WHERE cb.balance_raw != 0; -- Filter out 0 balances
"""
token_balances = flipside.query(sql)

#Format and create dataframe from flipside returned query
columns = token_balances.columns
rows = token_balances.rows
token_balance_logs_df = pd.DataFrame(rows, columns=columns)

# see df head
display(token_balance_logs_df.head())
```

```python
# Daily Balance Using ez _token_transfers

sql ="""
WITH airdrop_recipients AS (
    -- Base: wallets that claimed ARB airdrop
    SELECT 
        to_address AS wallet_address,
        block_timestamp AS claim_time,
        amount AS claimed_amount
    FROM arbitrum.core.ez_token_transfers
    WHERE 
        from_address = LOWER('0x67a24CE4321aB3aF51c2D0a4801c3E111D88C9d9')
        AND contract_address = LOWER('0x912CE59144191C1204E64559FE8253a0e49E6548')
        AND amount > 0
),

erc20_transfers AS (
    -- Extract only relevant ERC-20 transfers
    SELECT
        tt.block_number,
        tt.block_timestamp,
        tt.contract_address AS token_address,
        tt.from_address,
        tt.to_address,
        tt.amount AS amount_raw
    FROM arbitrum.core.ez_token_transfers tt
    JOIN arbitrum.core.dim_contracts tc 
        ON tt.contract_address = tc.address
    WHERE tt.contract_address = lower('0x912CE59144191C1204E64559FE8253a0e49E6548')
        and tt.amount > 0
),

filtered_transfers AS (
    -- Join transfers to recipients, apply 30-day limit
    SELECT
        t.block_number,
        t.block_timestamp,
        t.token_address,
        r.wallet_address,
        --handle both from and to transaction for wallet (from gets negative)
        CASE 
            WHEN t.from_address = r.wallet_address THEN -t.amount_raw
            WHEN t.to_address = r.wallet_address THEN t.amount_raw
            ELSE 0 -- Ignore other wallets
        END AS delta_raw
    FROM erc20_transfers t
    JOIN airdrop_recipients r
        ON (t.from_address = r.wallet_address OR t.to_address = r.wallet_address)
        AND t.block_timestamp >= r.claim_time --transfer on or after claim day
        AND t.block_timestamp < TIMESTAMPADD(DAY, 30, r.claim_time) -- transfer within 30 days of claim day
),

aggregated_balances AS (
    -- Aggregate net balances per wallet/token
    SELECT
        wallet_address,
        token_address,
        SUM(delta_raw) AS balance_raw,
        MAX(block_number) AS last_updated_block,
        MAX(block_timestamp) AS last_updated_timestamp
    FROM filtered_transfers
    GROUP BY wallet_address, token_address
    HAVING balance_raw != 0
)

SELECT
    ab.wallet_address,
    ab.token_address,
    tc.symbol AS token_symbol,
    tc.decimals,
    ab.balance_raw,
    ab.balance_raw / POWER(10, tc.decimals) AS balance_adjusted,
    ab.last_updated_block,
    ab.last_updated_timestamp
FROM aggregated_balances ab
JOIN arbitrum.core.dim_contracts tc 
    ON ab.token_address = tc.address;

"""
token_balances = flipside.query(sql)

#Format and create dataframe from flipside returned query
columns = token_balances.columns
rows = token_balances.rows
token_balance_logs_df = pd.DataFrame(rows, columns=columns)

# see df head
display(token_balance_logs_df.head())
```

```python
# Daily Balance Using ez _token_transfers

sql ="""
WITH airdrop_recipients AS (
    SELECT 
        to_address AS wallet_address,
        block_number AS claim_block,
        block_timestamp AS claim_time
    FROM arbitrum.core.ez_token_transfers
    WHERE 
        from_address = LOWER('0x67a24CE4321aB3aF51c2D0a4801c3E111D88C9d9')
        AND contract_address = LOWER('0x912CE59144191C1204E64559FE8253a0e49E6548')
        AND amount > 0
),

filtered_transfers AS (
    -- Use ez_token_transfers (MUCH smaller)
    SELECT
        ett.block_number,
        ett.block_timestamp,
        ett.contract_address AS token_address,
        ett.from_address,
        ett.to_address,
        ett.amount AS amount_raw,
        r.wallet_address,
        r.claim_time
    FROM arbitrum.core.ez_token_transfers ett
    JOIN airdrop_recipients r
        ON (ett.from_address = r.wallet_address OR ett.to_address = r.wallet_address)
        -- Keep only transfers within 30 days window
        AND ett.block_timestamp >= r.claim_time -- post claim date
        AND ett.block_timestamp < TIMESTAMPADD(DAY, 30, r.claim_time) -- pre 30 days after claim date
    -- OPTIONAL: ARB Token only
    WHERE ett.contract_address = LOWER('0x912CE59144191C1204E64559FE8253a0e49E6548')
),

agg_balances AS (
    SELECT
        wallet_address,
        token_address,
        SUM(
            CASE 
                WHEN from_address = wallet_address THEN -amount_raw
                WHEN to_address = wallet_address THEN amount_raw
                ELSE 0
            END
        ) AS balance_raw,
        MAX(block_number) AS last_updated_block,
        MAX(block_timestamp) AS last_updated_timestamp
    FROM filtered_transfers
    GROUP BY wallet_address, token_address
    HAVING balance_raw != 0
)

SELECT
    ab.wallet_address,
    ab.token_address,
--    tc.symbol AS token_symbol,
--    tc.decimals,
    ab.balance_raw,
--    ab.balance_raw / POWER(10, tc.decimals) AS balance_adjusted,
    ab.last_updated_block,
    ab.last_updated_timestamp
FROM agg_balances ab
--JOIN arbitrum.core.dim_contracts tc 
--    ON ab.token_address = tc.address;
"""
token_balances = flipside.query(sql)

#Format and create dataframe from flipside returned query
columns = token_balances.columns
rows = token_balances.rows
token_balance_logs_df = pd.DataFrame(rows, columns=columns)

# see df head
display(token_balance_logs_df.head())
```
